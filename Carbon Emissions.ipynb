```r
#Loading Dataset for Country Dataset
library(tidyverse)
CO2 = read.csv('shared_documents/FFG_Hackathon_Country_Level_Data.csv')
head(CO2)
```

```r
#Filtering columns out by the country name and year to find the amount of carbon emissions that existed in the timeframe
Carbon_Ozone = CO2 %>% group_by(Country.Name, Year) %>% summarise(Frequency = sum(EN.CO2.ETOT.ZS,na.rm=TRUE))
Carbon_Ozone
```

```r
#Wrote this into a CSV file as a summary table for more analysis
write.csv(Carbon_Ozone, file = "Total_Carbon_Emissions_By_Country.csv", row.names = FALSE)
```
